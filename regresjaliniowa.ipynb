# Regresja liniowa

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
no_samples = 100
area = np.random.random(no_samples) * 100 + 20
```

```python
area
```

```python
price = 6 * area + 10 + np.random.randn(no_samples) * 10
```

```python
plt.plot(area, price, '.')
plt.xlabel('Area')
plt.ylabel('Price [1000 PLN]')
plt.grid()
```

```python
M = np.array([
    [area[0], 1],
    [area[1], 1]
])
v = np.array([price[0], price[1]])
v
```

```python
print(f'det(M) = {np.linalg.det(M)}')
M_inv = np.linalg.inv(M)
u = np.matmul(M_inv, v)
u
```
